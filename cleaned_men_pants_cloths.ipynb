```python
import shutil
import os

# Original source path (make sure this matches your actual structure)
source_path = os.path.join(path, 'pants_dataset', 'pants_dataset', 'men_pants', 'cloths')

# Destination path in /content
destination_path = '/content/cleaned_men_pants_cloths'

# Copy the folder
shutil.copytree(source_path, destination_path)

print(f"✅ تم نسخ مجلد 'cloths' إلى: {destination_path}")
```

```python
import os

# List of image filenames to delete
images_to_delete = ['menpant2377.png', 'menpant253.png', 'menpant3229.png',
                    'menpant3375.png','menpant3480.png','menpant3801.png',
                    'menpant4351.png','menpant523.png','menpant5292.png',
                    'menpant6224.png','menpant6288.png','menpant6619.png',
                    'menpant6748.png','menpant6776.png','menpant782.png',
                    'menpant7989.png','menpant8033.png','menpant813.png',
                    'menpant889.png','menpant1045.png','menpant945.png',
                    'menpant9474.png','menpant9551.png','menpant9635.png',
                    'menpant966.png']  # Replace with actual names

# Path to the folder containing the images
folder_path = '/content/cleaned_men_pants_cloths'

# Loop through and delete
for image_name in images_to_delete:
    image_path = os.path.join(folder_path, image_name)
    if os.path.exists(image_path):
        os.remove(image_path)
        print(f"🗑️ Deleted: {image_name}")
    else:
        print(f"⚠️ Not found: {image_name}")
```

```python
import shutil

# Folder to zip
folder_to_zip = '/content/cleaned_men_pants_cloths'
# Output zip file path (without extension)
output_zip = '/content/cleaned_men_pants_cloths_zip'

# Create the zip file
shutil.make_archive(output_zip, 'zip', folder_to_zip)

print("✅ Folder zipped successfully.")
```
